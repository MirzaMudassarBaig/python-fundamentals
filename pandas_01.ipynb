```python
import pandas as pd
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

# Load the Excel files
df_drug = pd.read_excel("C:/Users/Dell/Desktop/Cinnamomum verum/drug_targets.xlsx", sheet_name="target1")
df_disease = pd.read_excel("C:/Users/Dell/Desktop/Cinnamomum verum/targets.xlsx", sheet_name="target2")

# Convert the relevant columns to sets (drop NaN and ensure values are strings)
set_drug = set(df_drug['Drug Targets'].dropna().astype(str))
set_disease = set(df_disease['Disease Targets'].dropna().astype(str))

# Plot the Venn diagram
plt.figure(figsize=(8, 8))
venn = venn2([set_drug, set_disease], set_labels=('Drug Targets', 'Disease Targets'))

# Customize Venn diagram labels
venn.get_label_by_id('10').set_text(f'Only in Drug Targets\n({len(set_drug - set_disease)})')
venn.get_label_by_id('01').set_text(f'Only in Disease Targets\n({len(set_disease - set_drug)})')
venn.get_label_by_id('11').set_text(f'Common\n({len(set_drug & set_disease)})')

# Add title
plt.title("Venn Diagram: Drug Targets vs Disease Targets")

# Display the plot
plt.show()
```

```python
import pandas as pd
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

# Read data from Excel files
data1 = pd.read_excel("C:/Users/Dell/Desktop/Cinnamomum verum/drug_targets.xlsx", sheet_name="target1")
data2 = pd.read_excel("C:/Users/Dell/Desktop/Cinnamomum verum/targets.xlsx", sheet_name="target2")

# Extract the relevant columns as sets
set1 = set(data1['Drug Targets'].dropna().astype(str))  # Ensure no NaN values and convert to string
set2 = set(data2['Disease Targets'].dropna().astype(str))

# Check lengths and sets
print(f"Set1 (Drug Targets): {len(set1)} items")
print(f"Set2 (Disease Targets): {len(set2)} items")

# Create the Venn diagram
plt.figure(figsize=(8, 8))
venn = venn2([set1, set2], set_labels=('Meningitis', 'Cinnamomum verum'))

# Customize diagram (optional)
venn.get_label_by_id('10').set_text(f'Only in Drug Targets: {len(set1 - set2)}')
venn.get_label_by_id('01').set_text(f'Only in Disease Targets: {len(set2 - set1)}')
venn.get_label_by_id('11').set_text(f'Common: {len(set1 & set2)}')

# Add custom title (optional)
plt.title("Venn Diagram of Drug and Disease Targets")

# Show the Venn diagram
plt.show()
```
