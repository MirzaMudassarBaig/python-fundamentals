# 16- spliting a string into multiple columns

```python
# import libraries
import pandas as pd 
df= pd.DataFrame({'name':['sana sani','ayesha ali','sara khan','saher arroj'],
                  'location':['lahore,pakistan','sargoda,pakistan','karachi,pakistan','islamabad,pakistan']})
df
```

```python
# spliting a column into two columns
df.name.str.split(' ',expand=True)
```

```python
# adding those splits into new columns
df[["first_name", "last_name"]]=df.name.str.split(' ',expand=True)
df
```

```python
print(df.head())
```

```python
df[['city', 'country']] = df['location'].str.split(',', expand=True)
```

```python
df
```

```python
# refine data manipulation
df= df[['first_name','last_name','city','country']]
df
```
